Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.tensor([1] * 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
vert = torch.cat([torch.cat([torch.arange(2, 10, 2).reshape(2, -1), torch.zeros(2, 3)], dim=1), torch.tensor([1] * 10).reshape(2, -1)], dim=0).int()
vert
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.tensor([1] * 10).reshape(2, -1)

Желаемый результат: новый тензор размером $2 \times 10$.

```python
gor = torch.cat([torch.cat([torch.arange(2, 10, 2).reshape(2, -1), torch.zeros(2, 3)], dim=1), torch.tensor([1] * 10).reshape(2, -1)], dim=1).int()
gor
```

Задание 23. Вставить в двумерный терзор из задания 17 строку из минимального элемента перед первой строкой.

```python
min_value = vert.min()
new_row = torch.full((1, 5), min_value)
new_vert = torch.cat([new_row, vert], dim=0)
```

Задание 24.	В тензоре, полученном при выполнении задания 23, каждый элемент заменить его корнем квадратным.

```python
for i in range(len(new_vert)):
    for j in range(len(new_vert[i])):
        new_vert[i][j] = new_vert[i][j] * new_vert[i][j]
new_vert        
```

Задание 25.	Заменить в тензоре из задания 18 все значения, кратные 3 или 2 на число 1, а остальные – оставить без изменений.

```python
for i in range(len(gor)):
    for j in range(len(gor[i])):
        if gor[i][j] % 2 == 0 or gor[i][j] % 3 == 0:
            gor[i][j] = 1
gor            
```
